```python
import pandas as pd

#importing the data
data1 = pd.read_csv("data/best-selling-books.csv")# "Book" e "Author"
data2 = pd.read_csv("data/BooksDatasetClean.csv") #"Title" e "Authors"
data3 = pd.read_csv("data/data.csv")#"title" e "authors"
data4 = pd.read_csv("data/GoodReads_100k_books.csv")#"author" e "title"
data5 = pd.read_csv("data/books_data.csv")#"Title" "authors"
reviews = pd.read_csv("data/Books_rating.csv")#Title

data1
```

```python
#cleaning the data before merging it (for each dataset)
data1 = data1.drop(columns=['Original language', 'First published','Approximate sales in millions' ])
data2 = data2.drop(columns=['Publisher', 'Price Starting With ($)','Publish Date (Month)', 'Publish Date (Year)' ])
data3 = data3.drop(columns=['isbn13', 'isbn10', 'subtitle', 'thumbnail', 'published_year', 'average_rating',	'ratings_count' ])
data4 =  data4.drop(columns=['bookformat', 'img', 'isbn', 'isbn13', 'link', 'rating', 'reviews', 'totalratings' ])
data5 = data5.drop(columns=['image', 'previewLink', 'publisher', 'publishedDate', 'infoLink', 'ratingsCount'])

#here data was inserted with [] for these two columns so i modified it
data5['authors'] = data5['authors'].str.strip('[]').str.replace("'", "")
data5['categories'] = data5['categories'].str.strip('[]').str.replace("'", "")

#handling the helpfulness of the reviews to choose which one to keep before dropping the column
reviews['helpfulness'] = reviews['review/helpfulness'].str.split('/').str[1].astype(int)
reviews['review'] = reviews['review/helpfulness'].str.split('/').str[0].astype(int)
#only rows where the score is above 50%
reviews = reviews[reviews['review'] / reviews['helpfulness'] > 0.5]
reviews = reviews.drop(columns=['Id', 'Price', 'User_id', 'profileName', 'review/helpfulness', 'review/score', 'review/time', 'helpfulness', 'review'])
```

```python
merged_1_2 = pd.merge(data1, data2, how='left', left_on='Book', right_on='Title') #Title Authors
merged_1_2['Authors'].fillna(merged_1_2['Author(s)'], inplace=True)
merged_1_2 = merged_1_2.drop(columns=['Authors', 'Category', 'Title'])
merged_1_2.rename(columns={'Book': 'Title', 'Author(s)':'Author'}, inplace=True)
merged_1_2
```

```python
merged_3_4 = pd.merge(data3, data4, how='outer', left_on='title', right_on='title')#title authors
merged_3_4['categories'].fillna(merged_3_4['genre'], inplace=True)
merged_3_4['authors'].fillna(merged_3_4['author'], inplace=True)
merged_3_4['num_pages'].fillna(merged_3_4['pages'], inplace=True)
merged_3_4['categories'].fillna(merged_3_4['genre'], inplace=True)
merged_3_4 = merged_3_4.drop(columns=['author', 'desc', 'genre', 'pages'])
merged_3_4.rename(columns={'title': 'Title', 'authors' : 'Author', 'categories' : 'Genre', 'description':'Description'}, inplace=True)
merged_3_4
```

```python
merged_first_four = pd.merge(merged_1_2, merged_3_4, how='outer', left_on='Title', right_on='Title') #Title Authors
merged_first_four['Author_x'].fillna(merged_first_four['Author_y'], inplace=True)
merged_first_four['Genre_x'].fillna(merged_first_four['Genre_y'], inplace=True)
merged_first_four['Description_x'].fillna(merged_first_four['Description_y'], inplace=True)
merged_first_four = merged_first_four.drop(columns=['Author_y', 'Genre_y', 'Description_y'])
merged_first_four.rename(columns={'Author_x': 'Author', 'Genre_x' : 'Genre', 'Description_x':'Description'}, inplace=True)

merged_first_four
```

```python
merge_all = pd.merge(merged_first_four, data5, how='outer', left_on='Title', right_on='Title') #Title Authors
merge_all['Author'].fillna(merge_all['authors'], inplace=True)
merge_all['Genre'].fillna(merge_all['categories'], inplace=True)
merge_all['Description'].fillna(merge_all['description'], inplace=True)
merge_all = merge_all.drop(columns=['description', 'authors', 'categories'])
merge_all.rename(columns={'Author_x': 'Author', 'Genre_x' : 'Genre', 'Description_x':'Description'}, inplace=True)

merge_all
```

```python
#handling duplicates
merge_all['nan'] = merge_all.isna().sum(axis=1)
duplicate_count = merge_all.duplicated(subset=['Title']).sum()
duplicated_rows = merge_all[merge_all.duplicated(subset=['Title'], keep=False)]
duplicated_rows
merge_all['nan'] = merge_all.isna().sum(axis=1)
lista = []
grouped_df = duplicated_rows.groupby('Title')
for title, group_df in grouped_df:
    min_nan_index = group_df['nan'].idxmin()
    lista.append(min_nan_index)

merge_all.drop(lista, inplace=True)

merge_all
```

```python
merge_all.to_csv('books_final.csv', index=False)
reviews.to_csv('reviews_final.csv', index=False)
```

```python
from collections import Counter
import string 

max_length = 1000000

def clean_and_get_best(text):
    #spacy necessary to remove abbreviations
    doc = nlp(text)
    text = ' '.join([token.lemma_ if token.lemma_ != "-PRON-" else token.text for token in doc])
    #tokenizing
    words = word_tokenize(text.lower()) 
    #stopwords removal
    stop_words = set(stopwords.words('english'))
    filtered_words = [word for word in words if word not in stop_words and word not in string.punctuation]
    #frequency calculation
    word_freq = Counter(filtered_words)
    sorted_words = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)
    top_twenty_words = [word for word, freq in sorted_words[:30]]
    return top_twenty_words


for index, row in reviewss.iterrows():
    review_text = row['review']
    filtered_words = []

    for start in range(0, len(review_text), max_length):
        chunk = review_text[start:start+max_length]
        filtered_words.extend(clean_and_get_best(chunk))
    
    reviewss.at[index, 'review'] = filtered_words
    print(index)

    
```
